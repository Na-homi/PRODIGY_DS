```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Melt the dataframe, keeping 'country name', 'country code', 'indicator name', and 'indicator code' as identifier variables
df_population = df.melt(id_vars=['Country Name', 'Country Code',
                                  'Indicator Name', 'Indicator Code'],
                        var_name='year', 
                        value_name='population')

# Check the first few rows of the melted data
df_population
```

**OBSERVATION**
- After melting the table displays population data for various countries and regions from the year 1960 to 2023. Each row represents a specific country or region, identified by its name and code, along with the total population for a given year. 

- The dataset contains 17,024 rows and 6 columns, including 'Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', 'year', and 'population'. The data shows a wide range of population sizes, from smaller countries like Aruba to larger regions like Africa Eastern and Southern.



```python
# Remove rows where 'year' or 'population' are NaN (missing)
df_population = df_population.dropna(subset=['year', 'population'])

# Remove rows where 'year' has incorrect column name like 'Unnamed: 68'
df_population = df_population[df_population['year'] != 'Unnamed: 68']

df_population
```

**OBSERVATION**

- After cleaning the data by removing rows where 'year' or 'population' are NaN and excluding rows with incorrect 'year' values like 'Unnamed: 68', the dataset now contains 16,930 rows and 6 columns. The table still includes population data for various countries and regions from 1960 to 2023, but with the erroneous and incomplete data removed, ensuring more accurate and reliable analysis.

**VIZUALIZATION**

- Bar Chart for categorical data(Country vs Population)

```python
# Select the first 10 rows for countries and their population
top_countries = df_population[['Country Name', 'population']].head(10)

# Create a bar chart
plt.figure(figsize=(10,6))
plt.bar(top_countries['Country Name'], top_countries['population'], color='skyblue')

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Distribution of Top 10 Countries')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

- Histogram for continuous data(Population by Year)

```python

# Extract population data for 2023
df_2023 = df_population[df_population['year'] == 2023]

# Plot histogram with log scale to handle the population variance
plt.figure(figsize=(10, 6))
plt.hist(df_2023['population'], bins=30, color='lightgreen', edgecolor='black')

# Mean and Median lines
mean_value = df_2023['population'].mean()
median_value = df_2023['population'].median()
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:,.0f}')
plt.axvline(median_value, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median_value:,.0f}')

# Labels and title
plt.xlabel('Population (log scale)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Population in 2023', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xscale('log')  # Applying log scale

# Show plot
plt.show()
```

- Most countries have relatively small populations, while a few giants like China and India push the numbers way up. That’s why the mean (328M) is way higher than the median (10.6M)—those outliers are doing some heavy lifting. 
- The log scale helps to show just how wide the gap is, making it clear that the majority of countries fall on the lower end of the population spectrum. In short, the world’s population isn’t evenly spread—it’s concentrated in a handful of highly populated nations, while most countries have much smaller numbers.
